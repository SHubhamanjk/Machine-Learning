```python
# prompt: read the csv which path is /content/dt reg data.PNG.csv and print the data

import pandas as pd
# The provided path is not valid for a CSV file.
# Did you mean to use a different path?

# Assuming you meant to read a CSV file named "dt reg data.PNG.csv" from the current directory:

data = pd.read_csv("/content/dt reg data.PNG.csv")
data
```

```python
x = data.loc[:, ['yt', 'guide']]
y = data.loc[:, ['result']]
```

```python
# prompt: split the x y into train and text dataset and text size should 0.1 and import decision tree regressor and then fit the train data and check the score on text data

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.1)

# Create a decision tree regressor
regressor = DecisionTreeRegressor()

# Fit the regressor to the training data
regressor.fit(X_train, y_train)

# Check the score on the test data
regressor.score(x, y)
# print(f"Score on test data: {score}")
```
