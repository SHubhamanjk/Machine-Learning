Simple Linear Regression :- y=mx+c

```python
from sklearn.datasets import make_regression
x,y = make_regression(n_features=1, noise=5, n_samples=10000,random_state=1)
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Ridge Regression :- Use to decrease the value of m that is weight .. it follow the SSC ( sum of square of coffiecient to perform the action ) rule to reduce the weight .. in this after increasing the value of alpha we see the decrease in weight but weight will aproch to 0 it will never be the exactly 0

```python
from sklearn.datasets import make_regression
x,y = make_regression(n_features=1, noise=5, n_samples=10000)
import matplotlib.pyplot as plt
plt.scatter(x,y,s=2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```python
from sklearn.linear_model import Ridge
reg = Ridge()
reg.fit(X_train,y_train)
```

```python
reg.score(X_test, y_test)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y_pred = reg.predict(x)
```

```python
plt.scatter(x,y,label='Actual',s=2)
plt.scatter(x,y_pred,label='Predicted',s=2)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
def play_with_ridge(alpha):
    reg = Ridge(alpha=alpha)
    reg.fit(X_train,y_train)
    print("Score is",reg.score(X_test, y_test),"\n")
    print("Coeff is ",reg.coef_,"\n")
    print("Intercept is ",reg.intercept_,"\n")
    y_pred = reg.predict(x)
    plt.scatter(x,y,label='Actual',s=2)
    plt.scatter(x,y_pred,label='Predicted',s=2)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.show()
```

```python
play_with_ridge(100000000000)
```

Lasso Regression :- This also do the same what ridge regression do but this follow SAC(sum of absolute value of coeff ) to reduce and it can be 0 after increasing the value of alpha

```python
from sklearn.linear_model import Lasso
# reg = Lasso()
```

```python
def play_with_lasso(alpha):
    reg = Lasso(alpha=alpha)
    reg.fit(X_train,y_train)
    print("Score is",reg.score(X_test, y_test),"\n")
    print("Coeff is ",reg.coef_,"\n")
    print("Intercept is ",reg.intercept_,"\n")
    y_pred = reg.predict(x)
    plt.scatter(x,y,label='Actual',s=2)
    plt.scatter(x,y_pred,label='Predicted',s=2)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.show()
```

```python
play_with_lasso(55)
```

```python
play_with_lasso(555)
```

```python
play_with_lasso(0.4)
```
