```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs, make_moons
x,y=make_blobs(n_features=2,n_samples=5000,centers=4)
```

```python
plt.scatter(x[:,0], x[:,1])
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(x)
```

```python
predict = kmeans.predict(x)
# plt.scatter(x[:,0], x[:,1])
plt.scatter(x[:,0], x[:,1],c=predict)
```

```python
x, y = make_moons(n_samples=1000, noise=0.06)
plt.scatter(x[:,0], x[:,1])
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(x)
predict = kmeans.predict(x)
plt.scatter(x[:,0], x[:,1],c=predict)
```

```python
from sklearn.cluster import AgglomerativeClustering
ag1 = AgglomerativeClustering(linkage='single')
ag1.fit(x)
plt.scatter(x[:,0], x[:,1],c=ag1.labels_)
```

```python
import matplotlib.pyplot as plt
import numpy as np
A = np.array([[5,3], [10,15],[15,12],[24,10],[30,30],[85,70],[71,80],[60,78],[70,55]])
plt.scatter(A[:,0],A[:,1])
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
X = linkage(A, 'centroid')
dendrogram(X)
plt.show()
```

```python
X = linkage(A, 'single')
dendrogram(X)
plt.show()
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
table = pd.DataFrame(iris.data,columns=iris.feature_names)
table.head()
```

```python
table['target'] = iris.target
table.head()
```

```python
x = table.drop(['target'], axis='columns')
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 10)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier1 = GaussianNB()
classifier1.fit(X_train, y_train)
```

```python
y_pred=classifier1.predict(X_test)
# y_pred
classifier1.score(X_test,y_test)
```

MUktinomial navie baiss used for categorical data contain frequcny , count type of stuff

```python
from sklearn.naive_bayes import MultinomialNB
classifier2 = MultinomialNB()
classifier2.fit(X_train, y_train)
```

```python
classifier2.score(X_test,y_test)
```

Bernauli NB is used for binarize data

```python
from sklearn.naive_bayes import BernoulliNB
classifier3 = BernoulliNB()
classifier3.fit(X_train, y_train)
classifier3.score(X_test,y_test)
```
