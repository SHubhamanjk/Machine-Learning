```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/DT data.PNG.csv')
data.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
data
```

```python
data.result.value_counts()
```

```python
entropy_results = -(5/10)*np.log2(5/10) -(5/10)*np.log2(5/10)
entropy_results
```

```python
data[data.yt == 1]
```

```python
entropy_results_yt_yes = -(5/6)*np.log2(5/6) -(1/6)*np.log2(1/6)
entropy_results_yt_yes
```

```python
data[data.yt == 0]
```

```python
yt_entropy = (6/10) * entropy_results_yt_yes +  (4/10) * 0
yt_entropy
```

```python
info_yt = entropy_results - (6/10) * entropy_results_yt_yes -  (4/10) * 0
info_yt
```

```python
data[data.guide == 1]
```

```python
entropy_results_guide_yes = -(4/5)*np.log2(4/5) -(1/5)*np.log2(1/5)
entropy_results_guide_yes
```

```python
data[data.guide == 0]
```

```python
entropy_results_guide_no = -(1/5)*np.log2(1/5) -(4/5)*np.log2(4/5)
entropy_results_guide_no
```

```python
guide_entropy = (5/10) * entropy_results_guide_yes +  (5/10) * entropy_results_guide_no
guide_entropy
```

```python
info_guide = entropy_results - guide_entropy
info_guide
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df = data
df.result = le.fit_transform(df.result)
df
```

```python
x = df[['yt','guide']]
y = df['result']
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
dt = DecisionTreeClassifier(criterion='entropy')
```

```python
dt.fit(x,y)
```

```python
gf=export_graphviz(dt,'dt.tree',feature_names=['yt','guide'])
gf
```

```python
dt.predict([[0,0]])
```

Decision Tree Classifier :- It spilit the data in diffrent levels according to parameters ..

if we go with entropy then we go with higher change in entropy between parent and child class ... it use the "-p(+ve)log(p)-p(-ve)log(p)" formula in backend to calculate entropy
