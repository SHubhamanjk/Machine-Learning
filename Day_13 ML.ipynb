```python
import pandas as pd
data = pd.read_csv('/content/pca.csv')
data.head()
```

```python
# prompt: now i wanna split data horizontally A to H coumns in X and I column in Y

X = data[['A','B','C','D','E','F','G','H']]
Y = data['I']
```

```python
# prompt: now i have to normalize the X using min max scaler

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
# prompt: split the data into train and test data using sklearn

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.1, random_state=12)
X_train
```

```python
# prompt: import SVC and fit train data

from sklearn.svm import SVC

svc_classifier = SVC()
svc_classifier.fit(X_train, y_train)
svc_classifier.score(X_test, y_test)
```

Hypertuniny :- By hypertuning we create a perfect model by finding the perfect combination and value of parameters

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
table = pd.DataFrame(iris.data,columns=iris.feature_names)
table.head()
```

```python
table['target'] = iris.target
table.head()
```

```python
x = table.drop(['target'], axis='columns')
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt1 = DecisionTreeClassifier(criterion= 'gini', splitter= 'best', max_depth=2)
dt2 = DecisionTreeClassifier(criterion= 'entropy', splitter= 'best', max_depth=5)
dt3 = DecisionTreeClassifier(criterion= 'gini', splitter= 'random', max_depth= 4)
dt4 = DecisionTreeClassifier(criterion= 'entropy', splitter= 'random', max_depth= 6)
```

```python
dt1.fit(X_train, y_train)
dt2.fit(X_train, y_train)
dt3.fit(X_train, y_train)
dt4.fit(X_train, y_train)
```

```python
print('dt1 -> ',dt1.score(X_test, y_test))
print('dt2 -> ',dt2.score(X_test, y_test))
print('dt3 -> ',dt3.score(X_test, y_test))
print('dt4 -> ',dt4.score(X_test, y_test))
```

Like this we can do hypertuniny manually

Someautomatic way for hypertuning

```python
dt = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
options = {'criterion': ('gini','entropy'),'splitter': ('best','random'),'max_depth' : list(range(1,7)) }
```

what does the gridsearchCv do ??

It use all possible combination of parameters and give us the best among them

```python
gs = GridSearchCV(dt,options)
gs.fit(X_train, y_train)
```

```python
# gs.cv_results_['params']
```

```python
gs.best_params_
```

```python
scores=gs.cv_results_['mean_test_score']
scores
```

```python
import numpy as np
np.argmax(scores)
```

```python
scores[2]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
dt = DecisionTreeClassifier()
```

```python
options = {'criterion': ('gini','entropy'),'splitter': ('best','random'),'max_depth' : list(range(1,7)) }
```

what does randomizedsearch CV do ??

It choose the parameters randomly and from them give us the best combo ... it may be type of luck game but it saves the computation time


```python
samples = 5
rs = RandomizedSearchCV(dt, param_distributions=options, n_iter=samples)
```

```python
rs.fit(X_train, y_train)
```

```python
rs.cv_results_['params']
```

```python
rs.best_params_
```

```python
scores = rs.cv_results_['mean_test_score']
scores
```

```python
np.argmax(scores)
```
